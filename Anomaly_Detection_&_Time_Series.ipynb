**Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

Answer:
Anomaly Detection (also known as Outlier Detection) is a machine learning and statistical technique used to identify data points, events, or patterns that significantly deviate from the expected behavior in a dataset. These anomalies often indicate critical incidents such as fraud, network intrusions, system failures, or medical irregularities.

In simple terms, anomaly detection helps detect data points that ‚Äúdon‚Äôt fit in‚Äù with the rest of the data. Detecting these irregularities is crucial in domains like banking (fraud detection), cybersecurity (intrusion detection), healthcare (disease outbreak detection), and manufacturing (fault detection in machinery).

Types of Anomalies

Anomalies can be categorized into three major types based on the context and data relationships:

1. Point Anomalies

A point anomaly occurs when a single data point is significantly different from all other data points in the dataset.
It‚Äôs the simplest and most common type of anomaly.

Example:
In a credit card transaction dataset, if most transactions are below $1,000 and suddenly there‚Äôs a transaction of $50,000, that transaction is a point anomaly.
Similarly, in temperature readings, a single day recording 80¬∞C when the average is around 25¬∞C would be an anomaly.

Use Case:
Fraud detection, malfunction detection in sensors, medical test abnormalities.

2. Contextual Anomalies

A contextual anomaly (also known as a conditional anomaly) is a data point that is normal in one context but abnormal in another.
This type of anomaly depends on contextual information, such as time, location, or environmental conditions.

Example:
A temperature of 30¬∞C is normal in summer but would be an anomaly in winter.
Similarly, a spike in network usage at midnight may be abnormal, while the same pattern might be normal during business hours.

Use Case:
Time-series data analysis (e.g., weather data, IoT sensor data, website traffic).

3. Collective Anomalies

A collective anomaly occurs when a group of data points, considered together, deviates significantly from the overall pattern, even though individual points may not appear abnormal on their own.

Example:
In network security, a sudden sequence of unusual packets may indicate a distributed denial-of-service (DDoS) attack, even though each packet by itself looks normal.
In stock trading, a series of rapid buy-sell transactions from a single account may indicate market manipulation.

Use Case:
Sequence or pattern-based anomaly detection such as network intrusion, system logs, and time-series pattern analysis.

**Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

Answer:
Anomaly detection can be approached in various ways depending on the nature of the data and the definition of what constitutes an ‚Äúoutlier.‚Äù Three widely used algorithms for this purpose are Isolation Forest, DBSCAN (Density-Based Spatial Clustering of Applications with Noise), and Local Outlier Factor (LOF). Each of these methods detects anomalies differently and is suited to specific types of data and problem domains.

1. Isolation Forest

The Isolation Forest algorithm is a tree-based ensemble method specifically designed for anomaly detection. It works on the principle that anomalies are ‚Äúfew and different‚Äù and can be isolated from the rest of the data points more easily than normal observations. The algorithm builds multiple random decision trees (isolation trees) by randomly selecting features and split values. Data points that require fewer splits to be isolated are considered anomalies, as they are located in sparse regions of the feature space.

Approach: Random partitioning of data into isolation trees. Points isolated in fewer steps are likely anomalies.

Advantages: Fast, efficient for high-dimensional datasets, and does not require distance or density estimation.

Use Cases: Financial fraud detection, network intrusion detection, manufacturing fault detection, and large-scale tabular datasets.

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

DBSCAN is a density-based clustering algorithm that groups together data points that are close to each other (high density regions) and marks points in low-density regions as anomalies or noise. Unlike traditional clustering methods such as K-Means, DBSCAN does not require specifying the number of clusters beforehand. It depends on two key parameters ‚Äî Œµ (epsilon), the neighborhood radius, and MinPts, the minimum number of points required to form a dense region. Points that do not belong to any cluster are considered anomalies.

Approach: Identifies clusters based on data density and classifies low-density points as outliers.

Advantages: Can detect arbitrarily shaped clusters, handles noise well, and does not assume data distribution.

Use Cases: Geospatial data analysis, anomaly detection in spatial patterns, and image segmentation tasks.

3. Local Outlier Factor (LOF)

The Local Outlier Factor algorithm measures the local deviation of a given data point‚Äôs density with respect to its neighbors. It compares the local density of each point with that of its nearest neighbors; if a point has a significantly lower density than its neighbors, it is considered an outlier. The LOF score quantifies this deviation ‚Äî higher scores indicate stronger anomalies.

Approach: Computes local density ratios among data points to identify samples that are isolated relative to their surroundings.

Advantages: Effective for datasets with varying density, and captures local rather than global anomalies.

Use Cases: Fraud detection, network intrusion detection, and environmental or sensor data with varying densities.

**Question 3: What are the key components of a Time Series? Explain each with one example.**

Answer:
A Time Series is a sequence of data points collected or recorded at successive points in time, often at uniform intervals. It is widely used in domains such as finance, economics, meteorology, healthcare, and manufacturing to analyze trends, make forecasts, and detect patterns over time. Every time series can be decomposed into several key components that help in understanding the underlying structure of the data. The main components of a time series are Trend, Seasonality, Cyclic variations, and Irregular (Random) components.

The Trend component represents the long-term movement or general direction of the data over an extended period. It shows whether the values are increasing, decreasing, or remaining stable over time. A trend captures the underlying growth or decline pattern that persists despite short-term fluctuations. For instance, the steady increase in global average temperatures over decades represents an upward trend due to climate change. Similarly, a company‚Äôs annual revenue that consistently grows year after year exhibits a positive trend.

The Seasonal component refers to repeating patterns or fluctuations that occur at fixed, regular intervals due to seasonal effects or periodic influences. These variations often occur within a single year, such as months, weeks, or quarters. Seasonality is common in data influenced by climate, holidays, or business cycles. For example, retail sales typically rise sharply during December due to holiday shopping, or electricity consumption tends to increase in summer because of higher air conditioning usage.

The Cyclic component represents fluctuations in the data that occur over periods longer than one year and are not fixed in duration. Unlike seasonality, which repeats at regular intervals, cycles are influenced by economic or business conditions and can vary in length and intensity. Cyclic behavior often corresponds to expansions and recessions in the economy. For example, stock market indices may rise and fall in multi-year cycles based on economic growth and downturns.

The Irregular or Random component consists of unpredictable, random variations that cannot be explained by trend, seasonality, or cycles. These fluctuations occur due to unforeseen events such as natural disasters, sudden political changes, or market shocks. For instance, an unexpected event like the COVID-19 pandemic caused abrupt changes in airline passenger data and stock prices ‚Äî a clear example of an irregular component.

In conclusion, understanding these components ‚Äî trend, seasonality, cyclic variations, and irregularity ‚Äî helps analysts decompose complex time series data into interpretable parts. This breakdown provides valuable insights into long-term behavior, periodic changes, and random disturbances, ultimately improving the accuracy of forecasting and decision-making.

**Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?**

Answer:
In time series analysis, a series is said to be stationary if its statistical properties such as mean, variance, and autocovariance remain constant over time. In other words, the behavior of a stationary time series does not depend on when it is observed. This means that the underlying data-generating process is stable and predictable across different time periods. Stationarity is a critical assumption for many time series models like ARIMA, as these models rely on consistent statistical patterns to make accurate forecasts.

A non-stationary time series, on the other hand, exhibits trends, seasonality, or changing variance over time. For example, the monthly sales of a company that steadily increase due to market growth show a non-stationary trend. Similarly, daily temperature data that follow a yearly seasonal cycle is also non-stationary because its mean changes periodically.

To determine whether a series is stationary, several statistical tests and visual methods are used. A simple visual inspection can reveal non-stationarity ‚Äî if the plot shows an upward or downward trend, or if the fluctuations increase or decrease over time, the series is likely non-stationary. A more formal approach is the Augmented Dickey-Fuller (ADF) test, which tests the null hypothesis that a unit root is present in the data (i.e., the series is non-stationary). If the test statistic is less than the critical value and the p-value is below a chosen significance level (e.g., 0.05), the null hypothesis is rejected, indicating that the series is stationary. Another commonly used method is the KPSS (Kwiatkowski-Phillips-Schmidt-Shin) test, which tests the opposite hypothesis ‚Äî that the data is stationary around a deterministic trend.

When a series is found to be non-stationary, it can be transformed into a stationary series using various techniques. The most common method is differencing, which involves subtracting the current observation from the previous one to remove trends and stabilize the mean. For instance, if

In some cases, logarithmic, square root, or Box-Cox transformations are used to stabilize the variance and reduce heteroscedasticity.

In conclusion, achieving stationarity is a fundamental step in time series modeling because it ensures that the relationships between observations remain consistent over time. Once the series is made stationary through differencing, detrending, or transformation, more reliable forecasting models such as ARIMA, SARIMA, or Exponential Smoothing can be effectively applied.

Time series forecasting involves building mathematical models that capture patterns such as trend, seasonality, and autocorrelation to make accurate predictions about future values. Among the most widely used models are the Autoregressive (AR), Moving Average (MA), Autoregressive Integrated Moving Average (ARIMA), Seasonal ARIMA (SARIMA), and Seasonal ARIMA with Exogenous variables (SARIMAX) models. Although they share similar foundations, each model differs in structure, assumptions, and scope of application.

The Autoregressive (AR) model explains the current value of a time series as a linear combination of its past values. It assumes that past observations have a direct influence on the present. The AR model of order p, denoted as AR(p), uses p previous observations for prediction. Mathematically, it is expressed as:

The ARIMA (Autoregressive Integrated Moving Average) model combines both AR and MA components and introduces differencing to handle non-stationarity. The model is denoted as ARIMA(p, d, q), where p is the number of autoregressive terms, d is the number of differences required to make the series stationary, and q is the number of moving average terms.
ARIMA is a versatile model suitable for non-seasonal data that shows trends or irregular patterns, such as monthly sales or economic indicators after differencing.

The SARIMA (Seasonal ARIMA) model extends ARIMA by including seasonal components to account for recurring patterns at fixed intervals. It incorporates additional parameters (P, D, Q, s), where s represents the length of the seasonal cycle. The full model is represented as SARIMA(p, d, q)(P, D, Q, s).

These external factors could include economic indicators, weather data, or marketing expenditures. By incorporating these variables, SARIMAX enhances predictive power, especially in complex real-world systems where external influences play a crucial role. For instance, SARIMAX can model sales data influenced not only by past trends and seasonality but also by advertising budgets or holidays.

**Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components**
(Include your Python code and output in the code box below.)

Answer:




```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.datasets import get_rdataset

# Step 2: Load the AirPassengers dataset
# Using an inbuilt dataset from statsmodels
data = get_rdataset('AirPassengers').data

# Convert to a time series with datetime index
data['Month'] = pd.date_range(start='1949-01-01', periods=len(data), freq='MS')
data.set_index('Month', inplace=True)
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Step 3: Plot the original time series
plt.figure(figsize=(10, 5))
plt.plot(data['Passengers'], label='Monthly Air Passengers', color='blue')
plt.title('Original AirPassengers Time Series (1949‚Äì1960)')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Decompose the time series
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative', period=12)

# Step 5: Plot decomposed components
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.show()
```

**Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)**

Answer:

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

# --------------------------------------------------------
# Step 2: Simulate or load a numerical dataset
# (In place of NYC Taxi Fare, we create a similar synthetic dataset)
# --------------------------------------------------------

# Generate synthetic data similar to taxi fare vs. distance
rng = np.random.RandomState(42)
X, _ = make_blobs(n_samples=500, centers=[[2, 20], [5, 50], [8, 80]], cluster_std=1.5, random_state=42)

# Add some artificial anomalies (extreme fares/distances)
outliers = rng.uniform(low=[0, 0], high=[10, 120], size=(20, 2))
X = np.vstack([X, outliers])

# Create DataFrame
data = pd.DataFrame(X, columns=['Trip_Distance', 'Fare_Amount'])

# --------------------------------------------------------
# Step 3: Apply Isolation Forest for anomaly detection
# --------------------------------------------------------
iso_forest = IsolationForest(contamination=0.05, random_state=42)
data['anomaly'] = iso_forest.fit_predict(data[['Trip_Distance', 'Fare_Amount']])

# Label anomalies (anomaly = -1)
anomalies = data[data['anomaly'] == -1]
normal = data[data['anomaly'] == 1]

# --------------------------------------------------------
# Step 4: Visualize anomalies on a 2D scatter plot
# --------------------------------------------------------
plt.figure(figsize=(8, 6))
plt.scatter(normal['Trip_Distance'], normal['Fare_Amount'], c='blue', label='Normal Data', alpha=0.6)
plt.scatter(anomalies['Trip_Distance'], anomalies['Fare_Amount'], c='red', label='Anomalies', marker='x', s=80)
plt.title('Anomaly Detection using Isolation Forest (Taxi Fare Data)')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.legend()
plt.grid(True)
plt.show()
```

**Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)**

Answer:

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.datasets import get_rdataset

# Step 2: Load the AirPassengers dataset
data = get_rdataset('AirPassengers').data
data['Month'] = pd.date_range(start='1949-01-01', periods=len(data), freq='MS')
data.set_index('Month', inplace=True)
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Step 3: Visualize the original time series
plt.figure(figsize=(10, 5))
plt.plot(data['Passengers'], label='Original Data', color='blue')
plt.title('Monthly AirPassengers Dataset (1949‚Äì1960)')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Fit the SARIMA model
# Seasonal period = 12 (monthly data)
model = sm.tsa.statespace.SARIMAX(
    data['Passengers'],
    order=(1, 1, 1),        # ARIMA parameters (p, d, q)
    seasonal_order=(1, 1, 1, 12),  # Seasonal parameters (P, D, Q, s)
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit(disp=False)

# Step 5: Forecast the next 12 months
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=forecast_steps, freq='MS')

# Step 6: Extract predicted mean and confidence intervals
pred_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Step 7: Plot forecast vs actual
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Passengers'], label='Observed', color='blue')
plt.plot(forecast_index, pred_mean, label='Forecast', color='orange')
plt.fill_between(forecast_index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='orange', alpha=0.2, label='95% Confidence Interval')
plt.title('SARIMA Forecast for AirPassengers (Next 12 Months)')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()

# Step 8: Print forecasted values
forecast_table = pd.DataFrame({
    'Forecast_Date': forecast_index,
    'Predicted_Passengers': pred_mean
})
print("Forecasted Passenger Counts for the Next 12 Months:")
print(forecast_table.round(2))
```

**Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)**

Answer:


```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# Step 2: Create a synthetic numerical dataset
# Generating clusters of normal data
X, _ = make_blobs(n_samples=400, centers=[[3, 3], [7, 7]], cluster_std=0.8, random_state=42)

# Add some artificial anomalies
rng = np.random.RandomState(42)
outliers = rng.uniform(low=0, high=10, size=(20, 2))
X = np.vstack([X, outliers])

# Convert to DataFrame
data = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])

# Step 3: Apply Local Outlier Factor
# n_neighbors controls the local region size for density comparison
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(data)

# -1 = anomaly, 1 = normal
data['anomaly'] = y_pred

# Step 4: Separate normal and anomalous data
normal_data = data[data['anomaly'] == 1]
anomalies = data[data['anomaly'] == -1]

# Step 5: Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(normal_data['Feature_1'], normal_data['Feature_2'], c='blue', label='Normal Data', alpha=0.6)
plt.scatter(anomalies['Feature_1'], anomalies['Feature_2'], c='red', marker='x', s=80, label='Anomalies')
plt.title('Anomaly Detection using Local Outlier Factor (LOF)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Print summary
print("Total Data Points:", len(data))
print("Number of Anomalies Detected:", len(anomalies))
```

SARIMA / SARIMAX)?
‚óè How would you validate and monitor the performance over time?

Answer:

Step 1: Real-Time Data Science Workflow Overview

In a real-world power grid scenario, energy consumption data arrives continuously from smart meters every 15 minutes.
The objective is twofold:

Forecast short-term energy demand to optimize generation and distribution.

Detect anomalies (spikes or drops) that may indicate faults, theft, or abnormal usage.

Step 2: Anomaly Detection Approach

Algorithm Choice:
Isolation Forest is preferred because it handles large, streaming numerical data efficiently and isolates outliers based on random partitioning.

Detects sudden spikes/drops in energy usage.

Works well in high-dimensional data (e.g., usage + weather).

Alternative:
Local Outlier Factor (LOF) could be used if the focus is on local density differences (useful for regional or cluster-based anomalies).

Step 3: Forecasting Model

Model Choice:
For short-term forecasting of energy demand, SARIMAX (Seasonal ARIMA with Exogenous variables) is ideal.

Captures trend, seasonality, and external factors like temperature or humidity.

Can forecast next few hours (e.g., next 4 √ó 15-minute intervals).

Step 4: Validation and Monitoring

Validation:
Use rolling cross-validation on time-based splits to simulate real-time conditions.

Monitoring:
Continuously compare predicted vs. actual energy usage.

Set dynamic thresholds for anomaly detection.

Log deviations and alert operators when consumption deviates significantly.

Step 5: Business Impact

This pipeline helps the power company:

Detect equipment faults or energy theft in real time.

Optimize load balancing by predicting upcoming demand.

Prevent power outages through proactive maintenance and anomaly alerts.

Improve energy pricing and scheduling using accurate forecasts.

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import statsmodels.api as sm

# Step 2: Simulate streaming energy consumption data (15-min intervals)
date_rng = pd.date_range(start='2023-01-01', end='2023-01-07', freq='15min')
np.random.seed(42)
energy_usage = 200 + 30*np.sin(np.linspace(0, 10*np.pi, len(date_rng))) + np.random.normal(0, 10, len(date_rng))

# Introduce artificial anomalies (valid indices within range)
energy_usage[100] = 400   # sudden spike
energy_usage[300] = 50    # sudden drop
energy_usage[500] = 380   # another spike

# Create DataFrame
data = pd.DataFrame({'Timestamp': date_rng, 'Energy_Usage': energy_usage})
data.set_index('Timestamp', inplace=True)

# Step 3: Anomaly Detection using Isolation Forest
iso_forest = IsolationForest(contamination=0.02, random_state=42)
data['Anomaly'] = iso_forest.fit_predict(data[['Energy_Usage']])

# Step 4: Visualize anomalies
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Energy_Usage'], color='blue', label='Energy Usage')
plt.scatter(data.index[data['Anomaly'] == -1],
            data['Energy_Usage'][data['Anomaly'] == -1],
            color='red', label='Anomalies', s=60)
plt.title('Real-Time Energy Consumption with Detected Anomalies')
plt.xlabel('Timestamp')
plt.ylabel('Energy Usage (kWh)')
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Train SARIMAX model for forecasting next 12 steps (3 hours ahead)
model = sm.tsa.statespace.SARIMAX(data['Energy_Usage'],
                                  order=(2, 1, 2),
                                  seasonal_order=(1, 1, 1, 96),  # 96 = 24*4 (daily seasonality)
                                  enforce_stationarity=False,
                                  enforce_invertibility=False)
results = model.fit(disp=False)

# Step 6: Forecast next 12 intervals
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1] + pd.Timedelta(minutes=15),
                               periods=forecast_steps, freq='15min')

# Step 7: Visualize forecast
pred_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

plt.figure(figsize=(12, 6))
plt.plot(data.index[-200:], data['Energy_Usage'][-200:], label='Observed', color='blue')
plt.plot(forecast_index, pred_mean, label='Forecast', color='orange')
plt.fill_between(forecast_index,
                 conf_int.iloc[:, 0],
                 conf_int.iloc[:, 1],
                 color='orange', alpha=0.2, label='Confidence Interval')
plt.title('Short-Term Energy Demand Forecast (Next 3 Hours)')
plt.xlabel('Time')
plt.ylabel('Energy Usage (kWh)')
plt.legend()
plt.grid(True)
plt.show()

# Step 8: Display forecasted values
forecast_table = pd.DataFrame({
    'Forecast_Timestamp': forecast_index,
    'Predicted_Energy_Usage': pred_mean
})
print("Forecasted Energy Usage for Next 3 Hours:")
print(forecast_table.round(2))
```
